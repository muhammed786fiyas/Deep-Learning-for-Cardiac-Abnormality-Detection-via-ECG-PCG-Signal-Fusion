```python
from scipy.io import loadmat
data = loadmat(r"E:\MAIN PROJECT\FILTERD CODES\data splitting\splitted data\PCG_split\a0064.mat")
ecg_signal = data['PCG_data']
```

```python
len(ecg_signal[0])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create time array
time = np.arange(0, len(ecg_signal[0]) / 2000, 1 / 2000)

# Plot ECG signal
plt.figure(figsize=(10, 6))
plt.plot(time, ecg_signal[0])
plt.xlabel('Time (s)')
plt.ylabel('PCG Signal')
plt.title('PCG Signal Plot')
plt.grid(True)
plt.show()
```
